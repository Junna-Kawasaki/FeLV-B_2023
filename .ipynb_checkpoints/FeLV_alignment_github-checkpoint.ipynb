# Sequence alignment for simplot++ analyses

```python
%%bash

mkdir output

##
cat \
    data/FeLV_ref.fasta | \
    seqkit replace -s -p "[^atcguATCGU]" -r "" \
    > output/FeLV_ref.QC.fasta

##
cat \
    data/FeLV-B.fasta | \
    seqkit replace -s -p "[^atcguATCGU]" -r "" \
    > output/FeLV-B.QC.fasta

##
mafft --auto \
    output/FeLV-B.QC.fasta \
    > output/FeLV-B.mafft.fasta

##
mafft --auto --keeplength \
    --addfragments output/FeLV_ref.QC.fasta \
    output/FeLV-B.mafft.fasta \
    > output/FeLV-B.mafft.smplot.fasta
```

# Phylogenetic analysis for the enFeLV derived region

## site 1-1192

```python
%%bash

cat output/FeLV-B.mafft.smplot.fasta | \
    seqkit subseq -r 1:1190 | \
    seqkit grep -n -r -p "AB674444" -v | \
    seqkit sort --by-name \
    > output/FeLV-B.mafft.selectB1.fasta
    
mafft \
    --auto \
    output/FeLV-B.mafft.selectB1.fasta \
    > output/FeLV-B.mafft.selectB1.aln.fasta
    
iqtree \
    -s output/FeLV-B.mafft.selectB1.aln.fasta \
    -m TESTNEW \
    -redo \
    -bb 1000 \
    -nt AUTO
```

# Phylogenetic analysis for the FeLV-A derived region

## site 1619-2550

```python
%%bash

cat output/FeLV-B.mafft.smplot.fasta | \
    seqkit subseq -r 1619:2550 | \
    seqkit grep -n -r -p "AB674444" -v | \
    seqkit grep -n -r -p "AY364318.1" -v | \
    seqkit sort --by-name --reverse  \
    > output/FeLV-B.mafft.selectA1.fasta

mafft --auto --keeplength \
    --addfragments data/FeLV-A.fasta \
    output/FeLV-B.mafft.selectA1.fasta \
    > output/FeLV-B.mafft.selectA1.aln.fasta
    
iqtree \
    -s output/FeLV-B.mafft.selectA1.aln.fasta \
    -m TESTNEW \
    -redo \
    -bb 1000 \
    -nt AUTO
```
