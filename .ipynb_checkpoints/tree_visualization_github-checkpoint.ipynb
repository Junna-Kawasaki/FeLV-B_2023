# enFeLV領域

```r
library("ggtree")

#input_f = "/Users/junna/Desktop/research/_2023/Nishigaki/230315_FeLVB/Onodera family FeLV seq data/FeLV-B.mafft.selectB.aln.fasta.treefile"
input_f = "/Users/junna/Desktop/research/_2023/Nishigaki/230406_FeLVB/新しいクローン名（小野寺）/FeLV-B.mafft.selectB.aln.fasta.treefile"

tree <- read.tree(input_f)
```

```r
ggtree(tree) + geom_text2(aes(subset=!isTip, label=node), hjust=-.3) + geom_tiplab()
```

```r
p <- ggtree(tree)
p1 <- p + geom_treescale(width=0.005) + geom_tiplab(size = 3) + 
    geom_nodelab(size = 3, col= "red", hjust=1.5, vjust=1.5) + xlim(0, 0.15)
p1
#viewClade(p1, 17) + geom_treescale(width=0.05) 
```
